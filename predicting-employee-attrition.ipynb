```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import LabelEncoder

import warnings
warnings.filterwarnings("ignore")

pd.set_option('display.max_columns', None)
```

```python
train_df= pd.read_csv("/kaggle/input/nodeb/train.csv")
test_df= pd.read_csv("/kaggle/input/nodeb/test.csv")
```

```python
test_df = test_df.drop(columns=["EmployeeCount", "EmployeeNumber", "StandardHours"],axis=1)
train_df = train_df.drop(columns=["EmployeeCount", "EmployeeNumber", "StandardHours"],axis=1)
```

```python
cat_variables=train_df.select_dtypes(include=['object']).columns
cat_variables
```

```python
label_encoder = LabelEncoder()
train_df[cat_variables] = train_df[cat_variables].apply(lambda column: label_encoder.fit_transform(column))
```

```python
test_df[cat_variables] = test_df[cat_variables].apply(lambda column: label_encoder.fit_transform(column))
```

```python
x = train_df.drop("Attrition", axis = 1)
y = train_df[["Attrition"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
log = LogisticRegression(random_state=42)
```

```python
param_grid = {
    'C': np.logspace(-4, 4, 20),  # Düzenleme parametresi için farklı değerler
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']  # Farklı optimizasyon algoritmaları
}

# Grid Search CV tanımla
grid_search = GridSearchCV(estimator=log, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2, scoring='accuracy')

# Grid Search ile modeli eğit
grid_search.fit(x_train, y_train)
```

```python
# En iyi parametreleri ve skoru yazdır
print("Best parameters:", grid_search.best_params_)
print("Best accuracy:", grid_search.best_score_)
```

```python
best_params = grid_search.best_params_
```

```python
#log = LogisticRegression(C= 0.08858667904100823, solver= 'newton-cg')
log = LogisticRegression(**best_params)

log.fit(x_train, y_train)
pred = log.predict(x_test)

# Model Evaluation
accuracy_test = accuracy_score(y_test, pred)
classification_rep_test = classification_report(y_test, pred)

print("Model: ",log)
print("")
print(f'Test Accuracy: {accuracy_test}')
print("")
print(f'Classification Report (Test Set):\n{classification_rep_test}')
```

```python
predicted_array = log.predict(test_df)

submission_df = pd.DataFrame(predicted_array)
```

```python
submission_df = pd.DataFrame(predicted_array, columns=["Attrition"])
submission_df['id'] = test_df.index
submission_df
```
